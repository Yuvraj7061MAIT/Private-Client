```python
from transformers import MBart50TokenizerFast, MBartForConditionalGeneration
```

```python
# Specify the model
model_name = "SnypzZz/Llama2-13b-Language-translate"
```

```python
# Load the tokenizer and model
tokenizer = MBart50TokenizerFast.from_pretrained(model_name, src_lang="en_XX")
model = MBartForConditionalGeneration.from_pretrained(model_name)
```

```python
# Define the English text to translate
text = "You love has made the person i am now"
```

```python
# Translate the text hindi
inputs = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"])
translated_text = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
# Translate the text Arabic
inputs = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id["ar_AR"])
translated_text1 = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
# Translate the text german
inputs = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id["de_DE"])
translated_text2 = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
# Translate the text spanish
inputs = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id["es_XX"])
translated_text3 = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```
